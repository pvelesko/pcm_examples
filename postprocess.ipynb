```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('counters.csv')
```

```python
plt.bar(data.index, data['time'], log=True)
plt.show()
```

```python
mean_time = data['time'].mean()
slow = data[data['time'] > mean_time]
fast = data[data['time'] <= mean_time]
slow = slow.sum() / len(slow)
fast = fast.sum() / len(fast)
```
